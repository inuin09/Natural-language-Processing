```python
#데이터 로딩
def read_data(filename, encoding='utf8'):  #읽기 함수 정의 
  with open(filename, 'r', encoding=encoding) as f:
    data =f.read() 
  return data

def write_data(data, filename, encoding='utf8'):  #쓰기 함수도 정의 
  with open(filename, 'w', encoding=encoding) as f:
    f.write(data)

data = read_data('김소월시.txt', encoding='utf8')  #전체 파일은 ratings.txt
```

```python
print(len(data))
```

```python
import rhinoMorph
rn = rhinoMorph.startRhino()
```

```python
#전체 데이터 형태소 분석
morphed_data = ''

for data_each in data:
  morphed_data_each = rhinoMorph.onlyMorph_list(rn, data, pos=['NNG','NNP','VV','VA','XR','IC','MM','MAG','MAJ'], eomi=True)
  joined_data_each= ' '.join([str(elem) for elem in morphed_data_each])
  if joined_data_each:
    morphed_data += data_each

#형태소 분석된 파일 저장
write_data(morphed_data, '김소월시.txt', encoding='utf8')
```

```python
#데이터 본문 출력
data = read_data('김소월시.txt', encoding='utf8')
print(len(data))

data_txt = data#데이터 본문
```

```python
#분석된 데이터 빈도 구하기
from collections import Counter

mergedText = data_txt.split()
print('mergedText :', mergedText)  #파일 나옴ㅇㅇ
type(mergedText)
```

```python
wordinfo = Counter(mergedText)  #하나의 리스트로 묶인 분리된 요소를 카운트함. (내림차순)
print('wordinfo :',wordinfo)
type(wordinfo)
```

```python
print('단어 총 빈도수 : ',sum(wordinfo.values()))
```

```python
#고빈도 단어 나타내기
def most_common(hist):
  t=[]
  for key, value in hist.items():
    t.append((value,key))
  t.sort(reverse=True)
  return t
```

```python
t=most_common(wordinfo)
print('김소월 고빈도 단어 :')
for freq, word in t[:10]:
  print(word,freq)
```

```python
#고빈도 단어비율

freq=[]
total_word = sum(wordinfo.values())

for v in wordinfo.values():
  freq.append(round((v/total_word)*100,2))

sortedFreq = sorted(freq, reverse=True)
print('김소월 고빈도 단어 비율 : ',sortedFreq)
```
